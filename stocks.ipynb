```python
from alpha_vantage.timeseries import TimeSeries
from alpha_vantage.techindicators import TechIndicators
import time
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
cases = pd.read_csv('https://covid.ourworldindata.org/data/ecdc/total_cases.csv')
usa_date = pd.DataFrame({"Date": cases["date"], "World": cases["World"], "USA": cases["United States"]})
usa_date['Date'] = usa_date.Date.astype(str)
w = 0
stock_list = ['ba', 'zm', 'ccl', 'goog', 'dji']
def new_stock(stock_sym):
    ts = TimeSeries(key=api_key, output_format='pandas')
    ti = TechIndicators(key=api_key, output_format='pandas')
    stock_sym2, meta_data = ts.get_daily(symbol=stock_sym, outputsize = 'full')
    stock_sym3 = pd.DataFrame({"Date": stock_sym2.index, eval("stock_sym"): stock_sym2["4. close"], eval("stock_sym") + "_Volume": stock_sym2["5. volume"] })
    stock_sym3.reset_index(drop=True, inplace=True)
    stock_sym3['Date'] = stock_sym3.Date.astype(str)
    return(stock_sym3)

for x in stock_list:
    w += 1
    print(w)
    if w == 1:
        first_frame = new_stock(x)
        time.sleep(10)
        print(first_frame)
    elif w == 2:
        all_frames = new_stock(x)
        fff = all_frames.merge(first_frame, left_on='Date', right_on='Date')
        print(fff)
        print("dig")
        time.sleep(10)
    elif w == 3:
        all_frames = new_stock(x)
        ggg = all_frames.merge(fff, left_on='Date', right_on='Date')
        print(ggg)
        time.sleep(10)
    elif w == 4:
        all_frames = new_stock(x)
        hhh = all_frames.merge(ggg, left_on='Date', right_on='Date')
        time.sleep(10)
    else:
        all_frames = new_stock(x)
        iii = all_frames.merge(hhh, left_on='Date', right_on='Date')
        jjj = iii.merge(usa_date, left_on='Date', right_on='Date')
        print(jjj)
        jjj.to_csv('dow_data.csv')
```
